```python
#Importing Libraries
import tkinter as tk
from tkinter import filedialog
from PIL import ImageTk, Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("archive/input_image.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(img)
plt.show()
print(cv2.__version__)
```

```python
gray_img =cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.axis("off")
plt.imshow(gray_img)
plt.show()
```

```python
inv_gray=cv2.bitwise_not(gray_img)
plt.axis("off")
plt.imshow(inv_gray)
plt.show(
```

```python
blur_img=cv2.GaussianBlur(inv_gray,(21,21),sigmaX=0,sigmaY=0)
plt.axis("off")
plt.imshow(blur_img)
plt.show()
```

```python
inv_blur_img=cv2.bitwise_not(blur_img)
plt.axis("off")
plt.imshow(inv_blur_img)
plt
```

```python
pencil_sketch=cv2.divide(gray_img,inv_blur_img,scale=256.0)
plt.axis("off")
plt.imshow(pencil_sketch)
plt.show()
```

```python
cv2.imwrite("archive/pencil_sketch.png",pencil_sketch)
```
